## Basis Vectors
- Can be scaled to represent any vector in a given vector space.
- Typically use unit vectors along axes of vector space.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Declaring Basis Vectors
- x1 = 1, y1 = 0
- x2 = 0, y2 = 1

```python
i = np.array([1, 0])
j = np.array([0, 1])
```

```python
plt.plot(i,  j)
```

### Scaling the Vector

```python
i_scaled = 1.5 * i
j_scaled = 2 * j
```

```python
plt.plot(i_scaled, j_scaled)
```
