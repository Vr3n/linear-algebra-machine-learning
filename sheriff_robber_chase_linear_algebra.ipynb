# Plotting a System of Linear Equations.

The linear equations can be plotted and visualized as a graph.

We will see how to plot Linear Equations using python.

## Robber and Sheriff Chase
- The Robber has robbed a bank and Sheriff has arrived at the crime location. Robber Left in a car that has a speed of 2.5km/min and has a lead of 5 minutes. Sheriff's car has a speed of 3.0km/min.
- Using linear algebra figure out how much time it will take the Sheriff to catch the Robber

#### Given
- Robber car speed = 2.5km/min
- Sheriff car speed = 3.0km/min
- Robber time is +5 mins.
We can write sheriff's equation as 3(t - 5)

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Creating X-Axis (Time taken to catch robber).
We will be using `np.linspace()` method to draw out the X-axis line space.
- The `linspace()` method takes parameters as `linspace(<start>, <finish>, <n_points>)`.
    - start = the number which the ___X___ axis should start (Origin).
    - finish = the max number to show on ___X___ axis.
    - n_points = Detail of the line.

```python
t = np.linspace(0, 40, 1000) # start, finish, n_points
```

### Formula for Distance Covered by Robber.

$$d_r = 2.5 * t$$

- ___d_r___ is distance travelled by the robber.
- ___2.5___ is the speed of the robbers car.
- ___t___ is the time taken.

```python
d_r = 2.5 * t
```

### Formula for Distance Covered by Sheriff.
$$d_s = 3 * (t - 5)$$
- ___d_s___ is distance travelled by the Sheriff.
- ___3___ is the speed of Sheriffs car.
- ___(t - 5)___ is time taken (We deduct 5 because sheriff arrived 5 minutes after robber ran away) 

```python
d_s = 3 * (t - 5)
```

## Plotting the Solution in Graph.
The criteria for visualization are as follows:
1. The title should be set as ___"Time taken to catch the Robber"___.
2. ___Y___ label should be set as ___"Distance (in kms)"___.
3. ___X___ label should be set as ___"Time Taken (in mins)"___.

```python
fig, ax = plt.subplots()
plt.title("Time Taken To Catch The Robber")
plt.xlabel("Time Taken (in mins)")
plt.ylabel("Distance (in kms)")
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, d_r, c="green")
ax.plot(t, d_s, c="blue")
plt.axvline(x=30, color="purple", linestyle="--")
_ = plt.axhline(y=75, color="purple", linestyle="--")
```
