## Linear Algebra 1 Exercise: The solar panel problem.

Jill designs solar panel as a hobby.
1. On ___April 1st___, Jills "Mark I" design begins generating ___1Kj/day___.
2. On ___May 1st___, Jills "Mark II" design begins generating ___4Kj/day___.

Answers to find:
1. What day is it when Jills Mark II design has generated as much total energy as Mark I design?
2. How much energy have both generated by the day?
3. What would the solution to (1.) & (2.) be if mark II design generated ___1Kj/day___.

### Solving 1: Total Equal energy of Mark I and Mark II design.
T = time

$$(T + 30) * 1 = 4T$$
$$T + 30 = 4T$$
$$T - 4T = -30$$
$$-3T = -30$$
$$T = \frac{-30}{-3}$$
$$T = 10$$


### Subsituting T with 10.
$$(10 + 30) * 1 = 4 * 10$$
$$40 = 40$$

### Solving in python.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 60, 1000)
```

```python
t_mark_I = t * 1
```

```python
t_mark_II = 4 * (t - 30)
```

```python
fig, ax = plt.subplots()
plt.title("Energy Generation of Both Models")
plt.xlabel("Time (in days)")
plt.ylabel("Energy (In Kj/day)")
ax.set_xlim([0, 60])
ax.set_ylim([0, 60])
ax.plot(t, t_mark_I, c="green")
ax.plot(t, t_mark_II,c="blue")
plt.axvline(x=40, color="purple", linestyle="--")
_ = plt.axhline(y=40, color="purple", linestyle="--")
```
